```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from xgboost.sklearn import XGBClassifier
import matplotlib.pyplot as plt

from util import load_data, fdr, plot_report
```

```python
x_train, x_test, x_oot, y_train, y_test, y_oot = load_data('var_30.csv')
x_train.head()
```

```python
dt = DecisionTreeClassifier(ccp_alpha=1e-07, 
                            max_depth=20, 
                            min_samples_leaf=64, 
                            class_weight=None,
                            criterion='gini')
```

```python
bt = XGBClassifier(n_estimators=200,
                   max_depth=5,
                   learning_rate=0.05,
                   subsample=0.5,
                   verbose=3, n_jobs=4)
```

```python
rf = RandomForestClassifier(n_estimators=300,
                            max_depth=17, 
                            max_features=10,
                            ccp_alpha=1e-6, 
                            n_jobs=4)
```

```python
clfs = [('decision tree', dt), ('random forest', rf), ('boosted tree', bt)]
```

```python
ensembel = VotingClassifier(clfs, voting='soft', n_jobs=2).fit(x_train, y_train)
```

```python
print(fdr(ensembel, x_train, y_train))
print(fdr(ensembel, x_test, y_test))
print(fdr(ensembel, x_oot, y_oot))
```
